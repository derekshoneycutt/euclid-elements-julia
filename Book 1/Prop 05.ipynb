# Book 1

## Proposition 5

In isoceles triangles the angles at the base are equal to one another, and, if the equal straight lines be produced further, the angles under the base will be equal to one another.

```julia
""" Construct an isosceles triangle """
function construct_isosceles(origin::Point2f, eq_side_dist, θ, rotationθ)
    B = origin + [eq_side_dist * cos(rotationθ), eq_side_dist * sin(rotationθ)]
    C = origin + [eq_side_dist * cos(θ + rotationθ), eq_side_dist * sin(θ + rotationθ)]
    [origin, Point2f(B), Point2f(C)]
end;
```

```julia
#setup the chart space...
f = Figure()
ax = Axis(f[1,1], title="Euclid's Elements Book I, Proposition 5")

#note: we will draw the point labels at the end for clarity, but they will be defined earlier

# Let ABC be an isosceles triangle having the side AB equal to the side AC;
d = rand(Uniform(0f0, 1f0))
θ = rand(Uniform(0f0, 1f0π))
A,B,C = construct_isosceles(Point2f0(0,0), d, θ, 0f0)
lines!([A, B], color=:blue)
lines!([A, C], color=:blue)
lines!([B, C], color=:blue)

# and let the straight lines BD, CE be produced further in a straight line with AB, AC. [Post. 2]
#               ASIDE: we will extend the same distance as the equal sides of the triangle, but could do random
extend = d #rand(Uniform(0f0, 1f0))
D = continue_line(A, B, extend)
E = continue_line(A, C, extend)
lines!([B, D], color=:purple)
lines!([C, E], color=:purple)

# I say that the angle ABC is equal to the angle ACB, and the angle CBD to the angle BCE.
# Let a point F be taken at random on BD;
F_at = rand(Uniform(0f0, extend))
F = continue_line(A, B, F_at)

# from AE the greater let AG be cut off equal to AF the less; [I. 3]
AG = draw_cut_line(C, E, B, F, color=:pink, linewidth=1f0, cursorlw=0.02f0)
fill_cut_line(AG)
G = Point(AG)

# and let the straight lines FC, GB be joined; [Post. 1]
lines!([F, C], color=:red)
lines!([G, B], color=:red)

# Then, since AF is equal to AG and AB to AC,
#   the two sides FA, AC are equal to the two sides GA, AB, respectively;
# and they contain a common angle, the angle FAG.

#   Therefore the base FC is equal to the base GB,
#   and the triangle AFC is equal to the triangle AGB.
# and the remaining angles will be equal to the remaining angles respectively,
# namely those which the equal sides subtend,
#   that is, the angle ACF to the angle ABG,
#   and the angle AFC to the angle AGB  [I.4]
AFC_AGB = compare_triangle(F, A, C, G, A, B, Point2f0(0, -1), 0f0, precision=5, cursorcolor=:green1, color=:pink, linewidth=5f0, cursorlinewidth=1f0)
fill_tricompare(AFC_AGB)

# And, since the whole AF is equal to the whole AG, and in these AB is equal to AC,
#   the remainder BF is equal to the remainder CG.
# But FC was also proved equal to GB;
#   therefore the two sides BF, FC are equal to the two sides CG, GB respectively;
# and the angle BFC is equal to the angle GCB,
# while the base BC is common to them;
# therefore the triangle BFC is also equal to the triangle CGB, and the remaining
# angles will be equal to the remaining angles respectively, namely those which
# the equal sides subtend;
#   therefore the angle FBC is equal to the angle GCB,
#   and the angle BCF to the angle CBG.
BFC_CGB = compare_triangle(F, B, C, G, C, B, Point2f0(1.5, -1), 0f0, precision=5, cursorcolor=:green4, color=:pink, linewidth=5f0, cursorlinewidth=1f0)
fill_tricompare(BFC_CGB)

# Accordingly, since the whole angle ABG was proved equal to the angle ACF,
# and in these angle CBG is equal to the angle BCF,
#   the remaining angle ABC is equal to the remaining angle ACB;
# and they are at the base of the triangle ABC.
# But the angle FBC was also proved equal to the angle GCB;
# and they are under the base.

# Therefore etc.
#       Q.E.D.



#Now draw the point labels
Apoint = text!(A, text="A")
Bpoint = text!(B, text="B")
Cpoint = text!(C, text="C")
Dpoint = text!(D, text="D")
Epoint = text!(E, text="E")
Fpoint = text!(F, text="F")
Gpoint = text!(G, text="G")

#Draw and show the chart :)
current_figure()
```

```julia
#setup the chart space...
f = Figure()
ax = Axis(f[1,1], title="Euclid's Elements Book I, Proposition 5")
xlims!(ax, -0.1,2)
ylims!(ax, -1.1,0.75)

#note: we will draw the point labels at the end for clarity, but they will be defined earlier

# Let ABC be an isosceles triangle having the side AB equal to the side AC;
#               ASIDE: We'll use the same points as above, but uncomment if wanting new points
#d = rand(Uniform(0f0, 1f0))
#θ = rand(Uniform(0f0, 1f0π))
#A,B,C = construct_isosceles(Point2f0(0,0), d, θ, 0f0)
lines!([A, B], color=:blue)
lines!([A, C], color=:blue)
lines!([B, C], color=:blue)

# and let the straight lines BD, CE be produced further in a straight line with AB, AC. [Post. 2]
#               ASIDE: we will extend the same distance as the equal sides of the triangle, but could do random
extend = d #rand(Uniform(0f0, 1f0))
D = continue_line(A, B, extend)
E = continue_line(A, C, extend)
BD = straight_line(B, D, color=:purple, linewidth=1f0, cursorwidth=0.01f0)
CE = straight_line(C, E, color=:purple, linewidth=1f0, cursorwidth=0.01f0)

# I say that the angle ABC is equal to the angle ACB, and the angle CBD to the angle BCE.
# Let a point F be taken at random on BD;
#F_at = rand(Uniform(0f0, extend))
F = continue_line(A, B, F_at)

# from AE the greater let AG be cut off equal to AF the less; [I. 3]
AG = draw_cut_line(C, E, B, F, color=:pink, linewidth=5f0, cursorlw=0.01f0)
G = Point(AG)

# and let the straight lines FC, GB be joined; [Post. 1]
FC = straight_line(F, C, color=:red, linewidth=1f0, cursorwidth=0.01f0)
GB = straight_line(G, B, color=:red, linewidth=1f0, cursorwidth=0.01f0)

# Then, since AF is equal to AG and AB to AC,
#   the two sides FA, AC are equal to the two sides GA, AB, respectively;
# and they contain a common angle, the angle FAG.

#   Therefore the base FC is equal to the base GB,
#   and the triangle AFC is equal to the triangle AGB.
# and the remaining angles will be equal to the remaining angles respectively,
# namely those which the equal sides subtend,
#   that is, the angle ACF to the angle ABG,
#   and the angle AFC to the angle AGB  [I.4]
AFC_AGB = compare_triangle(F, A, C, G, A, B, Point2f0(0, -1), 0f0, precision=5, cursorcolor=:green1, color=:pink, linewidth=5f0, cursorlinewidth=1f0)

# And, since the whole AF is equal to the whole AG, and in these AB is equal to AC,
#   the remainder BF is equal to the remainder CG.
# But FC was also proved equal to GB;
#   therefore the two sides BF, FC are equal to the two sides CG, GB respectively;
# and the angle BFC is equal to the angle GCB,
# while the base BC is common to them;
# therefore the triangle BFC is also equal to the triangle CGB, and the remaining
# angles will be equal to the remaining angles respectively, namely those which
# the equal sides subtend;
#   therefore the angle FBC is equal to the angle GCB,
#   and the angle BCF to the angle CBG.
BFC_CGB = compare_triangle(B, F, C, C, G, B, Point2f0(1.5, -1), 0f0, precision=5, cursorcolor=:green4, color=:pink, linewidth=5f0, cursorlinewidth=1f0)

# Accordingly, since the whole angle ABG was proved equal to the angle ACF,
# and in these angle CBG is equal to the angle BCF,
#   the remaining angle ABC is equal to the remaining angle ACB;
# and they are at the base of the triangle ABC.
# But the angle FBC was also proved equal to the angle GCB;
# and they are under the base.

# Therefore etc.
#       Q.E.D.



#Now draw the point labels
Apoint = text!(A, text="A")
Bpoint = text!(B, text="B")
Cpoint = text!(C, text="C")
DAlpha = Observable(0f0)
Dpoint = text!(D, text="D", color=@lift(RGBA(0.0, 0.0, 0.0, $DAlpha)))
EAlpha = Observable(0f0)
Epoint = text!(E, text="E", color=@lift(RGBA(0.0, 0.0, 0.0, $EAlpha)))
FAlpha = Observable(0f0)
Fpoint = text!(F, text="F", color=@lift(RGBA(0.0, 0.0, 0.0, $FAlpha)))
GAlpha = Observable(0f0)
Gpoint = text!(G, text="G", color=@lift(RGBA(0.0, 0.0, 0.0, $GAlpha)))

# Draw the animation!
framerate = 24
d_s = 24
t_rots = 6π
timestamps = range(0,t_rots, step=t_rots/(d_s*framerate))

record(f, "gifs/Euclid-I-05.01.gif", timestamps; framerate=framerate) do t

    # animate the drawing setup
    animate_line(BD, 0.5f0π, 1f0π, t)
    animate_line(CE, 1f0π, 1.5f0π, t)
    animate_cut_line(AG, 2f0π, 3f0π, t, fade_start=3.5f0π, fade_end=4.5f0π)
    animate_line(FC, 3f0π, 3.5f0π, t)
    animate_line(GB, 3.5f0π, 4f0π, t)


    # Animate showing the points D, E, F, G
    if t > 0.5π && t <= π
        showD = (t - 0.5π)/0.5π
        DAlpha[] = showD
    elseif t > π && t <= 1.5π
        DAlpha[] = 1f0
        showE = (t - π)/0.5π
        EAlpha[] = showE
    elseif t > 1.5π && t <= 2π
        DAlpha[] = 1f0
        EAlpha[] = 1f0
        showF = (t - 1.5π)/0.5π
        FAlpha[] = showF
    elseif t > 2π && t <= 3π
        DAlpha[] = 1f0
        EAlpha[] = 1f0
        FAlpha[] = 1f0
    elseif t > 3π && t <= 3.5π
        DAlpha[] = 1f0
        EAlpha[] = 1f0
        FAlpha[] = 1f0
        showG = (t - 3π)/0.5π
        GAlpha[] = showG
    elseif t > 3.5π
        DAlpha[] = 1f0
        EAlpha[] = 1f0
        FAlpha[] = 1f0
        GAlpha[] = 1f0
    end

    # animate the triangle comparisons
    animate_tricompare(AFC_AGB, 4f0π, 4.5f0π, t, fade_start=5π, fade_end=5.5π)
    animate_tricompare(BFC_CGB, 4.5f0π, 5f0π, t, fade_start=5.5π, fade_end=6π)
end;
```
