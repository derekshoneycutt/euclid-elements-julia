## Proposition 2

To place at a given point [as an extremity] a straight line equal to a given straight line.

```julia
#setup the chart space...
f = Figure()
ax = Axis(f[1,1], aspect=DataAspect(), title="Euclid's Elements Book I, Proposition 2")

#note: we will draw the point labels at the end for clarity, but they will be defined earlier

# Let A be a given point, and BC the given straight line
A = Point2f0(rand(Uniform(-1f0, 1f0)), rand(Uniform(-1f0, 1f0)))
B = Point2f0(rand(Uniform(-1f0, 1f0)), rand(Uniform(-1f0, 1f0)))
C = Point2f0(rand(Uniform(-1f0, 1f0)), rand(Uniform(-1f0, 1f0)))
BC = lines!([B,C], color=:blue)

# Thus it is required to place at the point A [as an extremity] a straight line equal to the given straight line BC.

# From the point A to the point B let the straight line AB be joined; [Post. 1]
AB = lines!([A,B], color=:red)
# and on it let the equilateral triangle DAB be constructed. [I. 1]
DAB = equilateral_triangle(A, B, color=:pink, linewidth=1f0, cursorlw=0.02f0)
fill_equilateral(DAB)
D = Point(DAB)

# Let the straight lines AE, BF be produced in a straight line with DA, DB; [Post. 2]
r_BC = norm(B-C)
E = continue_line(D, A, r_BC*1.5)
AE = lines!([A,E], color=:magenta)
F = continue_line(D, B, r_BC*1.5)
BF = lines!([B,F], color=:magenta)

# with centre B and distance BC let the circle CGH be described; [Post. 3]
r_CGH = norm(B-C)
CGH = lines!(Circle(B, r_CGH), color=:cyan)
#       ASIDE: H is an arbitrary spot on CGH, but G is an intersection point of CGH and BF
G = continue_line(D, B, r_CGH)
H = continue_line(B, D, r_CGH - norm(B-D))

# and again, with centre D and distance DG let the circle GKL be described. [Post. 3]
r_GKL = norm(D-G)
GKL = lines!(Circle(D, r_GKL), color=:gold)
#       ASIDE: Again, K is an arbitrary spot on GKL, L is the intersection of GKL and AE
L = continue_line(D, A, r_GKL - norm(D-A))
K = continue_line(B, D, r_GKL)

# Then, since the point B is the centre of the circle CGH,
#   BC is equal to BG.
# Again, since the point D is the centre of the circle GKL,
#   DL is equal to DG.
# And in these DA is equal to DB;
#   therefore the remainder AL is equal to the remainder BG [C.N. 3]

# But BC was also proved equal to BG;
#   therefore each of the straight lines AL, BC is equal to BG.
# And things which are equal to the same thing are also equal to one another; [C.N. 1]
#   therefore AL is also equal to BC.

# Therefore at the given point A the straight line AL is placed equal to the givne straight line BC.
AL = lines!([A,L], color=:green1)

# Being what it was required to do.


#Now draw the point labels, highlighting A and L!
Apoint = text!(A, text="A", color=:green1)
Bpoint = text!(B, text="B")
Cpoint = text!(C, text="C")
Dpoint = text!(D, text="D")
Epoint = text!(E, text="E")
Fpoint = text!(F, text="F")
Gpoint = text!(G, text="G")
Hpoint = text!(H, text="H")
Kpoint = text!(K, text="K")
Lpoint = text!(L, text="L", color=:green1)

#Draw and show the chart :)
current_figure()
```

```julia
#setup the chart space...
f = Figure()
ax = Axis(f[1,1], aspect=DataAspect(), title="Euclid's Elements Book I, Proposition 2")
xlims!(ax, -3.5,3.5)
ylims!(ax, -3.5,3.5)

#note: we will draw the point labels at the end for clarity, but they will be defined earlier

# Let A be a given point, and BC the given straight line
#           ASIDE : Will use same points defined above -- uncomment this definition if new points are needed
#A = Point2f0(rand(Uniform(-1f0, 1f0)), rand(Uniform(-1f0, 1f0)))
#B = Point2f0(rand(Uniform(-1f0, 1f0)), rand(Uniform(-1f0, 1f0)))
#C = Point2f0(rand(Uniform(-1f0, 1f0)), rand(Uniform(-1f0, 1f0)))
BC = lines!([B,C], color=:blue)

# Thus it is required to place at the point A [as an extremity] a straight line equal to the given straight line BC.

# From the point A to the point B let the straight line AB be joined; [Post. 1]
AB = straight_line(A, B, color=:pink, linewidth=5f0, cursorwidth=0.02f0)
# and on it let the equilateral triangle DAB be constructed. [I. 1]
DAB = equilateral_triangle(A, B, color=:pink, linewidth=5f0, cursorlw=0.02f0)
D = Point(DAB)

# Let the straight lines AE, BF be produced in a straight line with DA, DB; [Post. 2]
r_BC = norm(B-C)
E = continue_line(D, A, r_BC*1.5)
AE = straight_line(A, E, color=:pink, linewidth=5f0, cursorwidth=0.02f0)
F = continue_line(D, B, r_BC*1.5)
BF = straight_line(B, F, color=:pink, linewidth=5f0, cursorwidth=0.02f0)

# with centre B and distance BC let the circle CGH be described; [Post. 3]
r_CGH = norm(B-C)
CGH = whole_circle(B, r_CGH, vector_angle(B, C), color=:pink, linewidth=5f0)
#       ASIDE: H is an arbitrary spot on CGH, but G is an intersection point of CGH and BF
G = continue_line(D, B, r_CGH)
H = continue_line(B, D, r_CGH - norm(B-D))

# and again, with centre D and distance DG let the circle GKL be described. [Post. 3]
r_GKL = norm(D-G)
GKL = whole_circle(D, r_GKL, vector_angle(D, G), color=:pink, linewidth=5f0)
#       ASIDE: Again, K is an arbitrary spot on GKL, L is the intersection of GKL and AE
L = continue_line(D, A, r_GKL - norm(D-A))
K = continue_line(B, D, r_GKL)

# Then, since the point B is the centre of the circle CGH,
#   BC is equal to BG.
BC_BG = compare_lines(B, C, B, G, Point2f0(-2.5, -2.5), /2f0, precision=5, cursorcolor=:pink, color=:pink, cursorlinewidth=0.75f0, linewidth=5f0)
# Again, since the point D is the centre of the circle GKL,
#   DL is equal to DG.
DL_DG = compare_lines(D, L, D, G, Point2f0(-2.25, -2.5), /2f0, precision=5, cursorcolor=:pink, color=:pink, cursorlinewidth=0.75f0, linewidth=5f0)
# And in these DA is equal to DB;
DA_DB = compare_lines(D, A, D, B, Point2f0(-2, -2.5), /2f0, precision=5, cursorcolor=:pink, color=:pink, cursorlinewidth=0.75f0, linewidth=5f0)
#   therefore the remainder AL is equal to the remainder BG [C.N. 3]
AL = straight_line(A, L, color=:green1, linewidth=5f0, cursorwidth=0.02f0)
AL_BG = compare_lines(A, L, B, G, Point2f0(-1.75, -2.5), /2f0, precision=5, cursorcolor=:pink, color=:pink, cursorlinewidth=0.75f0, linewidth=5f0)

# But BC was also proved equal to BG;
#   therefore each of the straight lines AL, BC is equal to BG.
# And things which are equal to the same thing are also equal to one another; [C.N. 1]
#   therefore AL is also equal to BC.
AL_BC = compare_lines(A, L, B, C, Point2f0(-1.5, -2.5), /2f0, precision=5, cursorcolor=:pink, color=:pink, cursorlinewidth=0.75f0, linewidth=5f0)

# Therefore at the given point A the straight line AL is placed equal to the given straight line BC.

# Being what it was required to do.


#Now draw the point labels, highlighting A and L!
Apoint = text!(A, text="A", color=:green1)
Bpoint = text!(B, text="B")
Cpoint = text!(C, text="C")
LAlpha = Observable(0f0)
Lpoint = text!(L, text="L", color=@lift(RGBA(0.0, 0.8, 0.0, $LAlpha)))


# Draw the animation!
framerate = 24
d_s = 24
t_rots = 9
timestamps = range(0,t_rots, step=t_rots/(d_s*framerate))

record(f, "gifs/Euclid-I-02.01.gif", timestamps; framerate=framerate) do t

    # animate the drawing setup
    animate_line(AB, 0.5f0, 1f0, t, fade_start=2, fade_end=3)
    animate_equilateral(DAB, 1f0, 2f0, t, fade_start=7.5, fade_end=8)
    animate_line(AE, 2f0, 2.5f0, t, fade_start=7.5, fade_end=8)
    animate_line(BF, 2f0, 2.5f0, t, fade_start=7.5, fade_end=8)
    animate_circle(CGH, 2.5f0, 3.5f0, t, fade_start=7.5, fade_end=8)
    animate_circle(GKL, 3.5f0, 4.5f0, t, fade_start=7.5, fade_end=8)

    # Animate showing the point L
    if t > 4.5 && t <= 5
        showL = (t - 4.5)/0.5
        LAlpha[] = showL
    elseif t > 5
        LAlpha[] = 1f0
    end
    # Animate drawing AL
    animate_line(AL, 4.5f0, 5f0, t)

    # animate the line comparisons
    animate_linecompare(BC_BG, 5f0, 5.5f0, t, fade_start=8, fade_end=8.5)
    animate_linecompare(DL_DG, 5.5f0, 6f0, t, fade_start=8, fade_end=8.5)
    animate_linecompare(DA_DB, 6f0, 6.5f0, t, fade_start=8, fade_end=8.5)
    animate_linecompare(AL_BG, 6.5f0, 7f0, t, fade_start=8, fade_end=8.5)
    animate_linecompare(AL_BC, 7f0, 7.5f0, t, fade_start=8, fade_end=8.5)
end;
```
